```python
# Load the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read in both dataframes
aqi_17 = pd.read_csv("https://aqs.epa.gov/aqsweb/airdata/daily_aqi_by_county_2017.zip", compression = 'zip')
aqi_18 = pd.read_csv("https://aqs.epa.gov/aqsweb/airdata/daily_aqi_by_county_2018.zip", compression = 'zip')

# Concatenate the DataFrames aqi_17 and aqi_18 vertically
aqi = pd.concat([aqi_17, aqi_18])

# Simplify column names
aqi.columns = (aqi.columns.str.lower().str.replace(' ','_'))

# Select the data of interest from Santa Barbara county
aqi_sb = aqi[aqi['county_name'] == 'Santa Barbara']
aqi_sb = aqi_sb.drop(columns = ['state_name', 'county_name', 'state_code', 'county_code'])

# Convert the date column to 'pandas.Datetime' object and update the index to date column
aqi_sb['date'] = pd.to_datetime(aqi_sb['date'])
aqi_sb = aqi_sb.set_index('date')

# Calculate a rolling average over 5 days and mean AQI
rolling_average = aqi_sb['aqi'].rolling('5D').mean()
aqi_sb['five_day_average'] = aqi_sb['aqi'].rolling('5D').mean()

# Plot showing both the daily AQI and the 5-day average
daily_aqi_rollingavg = aqi_sb[['aqi', 'five_day_average']].plot(ylabel = 'Air Quality Index (AQI)',
                                        xlabel = 'Year',
                                        title = 'Daily AQI and the 5-day average in Santa Barbara County',
                                        color = ['green', 'blue'])
daily_aqi_rollingavg.legend(['Daily AQI', '5-Day Average'], title = 'Air Pollution Level')
```

In the next cell we will calculate an average over a [rolling window](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.rolling.html) using the `rolling()`method for `pandas.Series`:

- `rolling()` is a lazy method, so we need to specify what we want to calculate over each window before it does something. 
- in this example we use the aggregator function `mean()` to calculate the average over each window
- the parameter '5D' indicates we want the window for our rolling average to be 5 days. 
- we get a `pandas.Series` as ouput

Store your answer in the `rolling_average` variable. You should have two columns in your series, `date` and the averages.
